````python
These are excellent questions for analyzing a **Spotify-style music dataset**! Let's walk through how you can handle each of these tasks using Python and libraries like `pandas`, `seaborn`, and `matplotlib`.

Assuming your DataFrame is called `df`, here’s how to tackle each point:

---

### ✅ 1. **Read Data, Handle Nulls & Duplicates**
```python
import pandas as pd

df = pd.read_csv('your_file.csv')  # or .xlsx

# Nulls
print(df.isnull().sum())
df = df.dropna()  # or fillna() based on context

# Duplicates
df = df.drop_duplicates()
```

---

### 📊 2. **Distribution of Track Popularity (Histogram)**
```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['popularity'], bins=30, kde=True)
plt.title('Distribution of Track Popularity')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.show()
```

---

### 🔁 3. **Popularity vs Duration (Scatter Plot)**
```python
sns.scatterplot(data=df, x='duration_ms', y='popularity')
plt.title('Popularity vs Duration of Tracks')
plt.xlabel('Duration (ms)')
plt.ylabel('Popularity')
plt.show()
```

---

### 🎤 4. **Artist with Most Tracks (Countplot)**
```python
top_artists = df['artist_name'].value_counts().head(10)
sns.barplot(x=top_artists.values, y=top_artists.index)
plt.title('Top 10 Artists by Track Count')
plt.xlabel('Number of Tracks')
plt.ylabel('Artist')
plt.show()
```

---

### 🧊 5. **Top 5 Least Popular Tracks**
```python
least_popular = df.sort_values('popularity').head(5)[['artist_name', 'track_name', 'popularity']]
print(least_popular)
```

---

### 🏆 6. **Top 5 Most Popular Artists (Avg Popularity)**
```python
top_artists_by_popularity = df.groupby('artist_name')['popularity'].mean().sort_values(ascending=False).head(5)
print(top_artists_by_popularity)
```

---

### 🎶 7. **Most Popular Tracks by Top 5 Artists**
```python
top_5_artists = top_artists_by_popularity.index.tolist()

for artist in top_5_artists:
    top_track = df[df['artist_name'] == artist].sort_values('popularity', ascending=False).head(1)
    print(f"{artist}: {top_track['track_name'].values[0]} (Popularity: {top_track['popularity'].values[0]})")
```

---

### 🔗 8. **Pair Plot of Numerical Features**
```python
sns.pairplot(df[['popularity', 'duration_ms', 'danceability', 'energy', 'valence']])
plt.suptitle('Pair Plot of Track Features', y=1.02)
plt.show()
```

---

### ⌛ 9. **Duration Variation Across Artists (Box/Violin Plot)**
```python
top_artists_list = df['artist_name'].value_counts().head(5).index.tolist()
sns.violinplot(data=df[df['artist_name'].isin(top_artists_list)], x='artist_name', y='duration_ms')
plt.title('Track Duration by Artist')
plt.xlabel('Artist')
plt.ylabel('Duration (ms)')
plt.xticks(rotation=45)
plt.show()
```

---

### 📈 10. **Popularity Distribution by Artist (Swarm/Violin Plot)**
```python
sns.violinplot(data=df[df['artist_name'].isin(top_artists_list)], x='artist_name', y='popularity', inner='quartile')
plt.title('Popularity Distribution by Artist')
plt.xlabel('Artist')
plt.ylabel('Popularity')
plt.xticks(rotation=45)
plt.show()
```

---

Want me to assemble this into a **ready-to-run notebook or Python script** with titles and comments included? Just say the word!
````
